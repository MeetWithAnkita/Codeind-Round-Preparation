An e-commerce site has a series of advertisements to display. Links to the ads are stores in a data structure and they are displayed or not is based on the value at bit position in a number. The sequence of ads being displayed at this time can be represented as a binary value. Where 1 means the ad is displayed and 0 means ad is not displayed. The ads should rotate. So, when the next page loads, ads that are displayed now are hidden and vice versa.


Sample input: 50
Sample output: 13


50 in bit (110010) ==> (001101) = 13

```python
def chnageads(num):
    # No converted to binary number using bin() builting function
    binary_num = bin(num)
    # delete the '0b' prefix from the binary number
    binary_num = binary_num[2:]
    print(binary_num)
    # String is Immutable in Python, so can't modify, take a new emply string
    b1 = ''
    # 0 to 1 and 1 to 0
    for i in binary_num:
        if i == '0':
            b1 += '1'
        else:
            b1 += '0'
    print(b1)

    # binary no converted to decimal no
    sum = 0 
    for i in range(len(b1)):
        if b1[i] == '1':
            sum += 2 ** (len(b1) - 1 - i)  # 6th element= 0 ,i =0 ==>  element * 2^ (6-1-i)= 2^5 = 0 
    print(sum)


chnageads(50)
```

Write a python program to sort characters and numbers so that first alphabets and then numbers are printed.

Sample input:   A7B1R3
Sample Output:  ABR137

```python
def input(n):
    # emply list to store the characters and digits
    A= []
    N = []
    for i in n:
        if i.isalpha():
            A.append(i)
        elif i.isdigit():
            N.append(i)
        else:
            pass
    # sorted for both list
    print(sorted(A))
    print(sorted(N))

    # Combine the sorted lists
    result = sorted(A) + sorted(N)
    print(dtype()''.join(result))

input('A7B1R3')
```
