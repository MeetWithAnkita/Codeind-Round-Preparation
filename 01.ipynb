An e-commerce site has a series of advertisements to display. Links to the ads are stores in a data structure and they are displayed or not is based on the value at bit position in a number. The sequence of ads being displayed at this time can be represented as a binary value. Where 1 means the ad is displayed and 0 means ad is not displayed. The ads should rotate. So, when the next page loads, ads that are displayed now are hidden and vice versa.


Sample input: 50
Sample output: 13


50 in bit (110010) ==> (001101) = 13

```python
def chnageads(num):
    # No converted to binary number using bin() builting function
    binary_num = bin(num)
    # delete the '0b' prefix from the binary number
    binary_num = binary_num[2:]
    print(binary_num)
    # String is Immutable in Python, so can't modify, take a new emply string
    b1 = ''
    # 0 to 1 and 1 to 0
    for i in binary_num:
        if i == '0':
            b1 += '1'
        else:
            b1 += '0'
    print(b1)

    # binary no converted to decimal no
    sum = 0 
    for i in range(len(b1)):
        if b1[i] == '1':
            sum += 2 ** (len(b1) - 1 - i)  # 6th element= 0 ,i =0 ==>  element * 2^ (6-1-i)= 2^5 = 0 
    print(sum)


chnageads(50)
```

Write a program to display the words that are repeated more than or equal to N times in the text.


Input: cat batman latt matter cat matter cat cat latt latt ,  3


output: cat latt 

```python
str = 'cat batman latt matter cat matter cat cat latt latt'
count = 3 

def count_occurences(str, count):
    # split the string
    words = str.split() #convert the string into list of words
    print(words)

    # count the occurences of each word
    word_count = dict() # empty dictionary to store the count of each word
    for w in words:
        if w in word_count:
            word_count[w] = word_count[w] + 1 
        else:
            word_count[w] = 1 
    print(word_count) #dictionary with word as key and count as value

    # return the count of words that occur more than 'count' times
    final_words = []
    for key in word_count:
        if word_count[key] >= count:
            final_words.append(key)
    print(final_words)
    
count_occurences(str, count)
```

Write a program for finding pair of highest product.

Input Array: [5, 3, 1, 4, 3, 7, 6, 9, 2]


Maximum product pair is: (7, 9)

input: [0, -2, -1, -4, 5, -8, 7, -8, 8]


maximum product pair is: (-8, -8)

```python
# arr = [0, -2, -1, -4, 5, -8, 7, -8, 8]
arr = [5,3,1,4,3,7,6,9,2]
# arr = [2]
def max_product(arr):
    if len(arr) < 2:
        print("Have no pair, add numbers...")
    else:
        max_a = arr[0]
        max_b = arr[1]
        for i in range(0, len(arr)):
            for j in range(i+1, len(arr)):
                if arr[i]*arr[j] > max_a*max_b:
                    max_a = arr[i]
                    max_b = arr[j]
        print("(",max_a,",",max_b,")")

max_product(arr)
```

Given an array of integers nums' and an integer 'target return indices of two numbers such that they add up to target.
You may assume that each input would have exactly one solution, and you may not use the same element twice.
You can return answer in any order.

ex1.
Input: nums = [2,7,11,15],
target = 9
Output: [0,1]
exp: nums[0]+nums[1]==target


ex2. nums [3,2,4], target = 6
Output: [1,2]


ex3. nums [3,3], target = 6
Output: [0,1]

```python
# nums = [2,7,11,15], 9
nums = [3,2,4]
target = 6
def solution(nums, target):
    for i in range(len(nums)):
        for j in range (i+1, len(nums)):
            if target == (nums[i] + nums[j]):
                print('[',i,',',j,']')

solution(nums, target)
```
